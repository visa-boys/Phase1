```python
room0 = data['Room 0 Area (sq. meters)']
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
room0.plot.hist(bins=60)
plt.title("Room 0 distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, lognorm, poisson, gamma, beta, weibull_min
from scipy.optimize import curve_fit

data1= room0 

fit_functions = [norm, expon, lognorm, gamma, beta, weibull_min]

for dist_func in fit_functions:
    params = dist_func.fit(data1)
    fitted_distribution = dist_func(*params)

    x = np.linspace(min(data), max(data), 1000)
    plt.hist(data1, bins=30, density=True, alpha=0.5, label='Data')
    plt.plot(x, fitted_distribution.pdf(x))
8
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(5, 2, figsize=(12, 18))
axes = axes.ravel()  # Flatten the 2D array for easy iteration

for i, room_name in enumerate(rooms_list):
    var = data[room_name]
    axes[i].hist(var, bins=20)
    axes[i].set_title("Area of " + room_name)

# Remove any extra empty subplots (if you have fewer than 10 rooms)
for j in range(len(rooms_list), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
